```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('problem1.csv')
df2 = pd.read_csv('problem2.csv')
df3 = pd.read_csv('problem3.csv')
df4 = pd.read_csv('problem4.csv')
```

```python
df = pd.concat([df1, df2, df3, df4])
```

```python
search_algos = ['breadth_first_search', 'depth_first_graph_search', 'uniform_cost_search', 
                'greedy_best_first_graph_search with h_unmet_goals', 'greedy_best_first_graph_search with h_pg_levelsum',
                'greedy_best_first_graph_search with h_pg_maxlevel', 'greedy_best_first_graph_search with h_pg_setlevel',
                'astar_search with h_unmet_goals', 'astar_search with h_pg_levelsum',
                'astar_search with h_pg_maxlevel', 'astar_search with h_pg_setlevel']
algos_index = list(range(1,12))
```

```python
df.replace({'Problem':{1:'P1', 2:'P2', 3:'P3', 4:'P4', 5:'P5'}}, inplace=True)
```

```python
fig, ax = plt.subplots()
for key, grp in df.groupby(['Search']):
    ax = grp.plot(ax=ax, kind='line', x='Actions', y='Expansions', label=key, logy=True)#, c=key, label=key)
plt.legend(bbox_to_anchor=(1,1))
#plt.legend(loc='best')
```

```python
fig, ax = plt.subplots()
for key, grp in df.groupby(['Search']):
    ax = grp.plot(ax=ax, kind='line', x='Actions', y='Time', label=key, logy=True)#, c=key, label=key)
#plt.ylim([0,110])
plt.legend(bbox_to_anchor=(1,1))
```

```python
fig, ax = plt.subplots()
for key, grp in df.groupby(['Search']):
    ax = grp.plot(ax=ax, kind='line', x='Actions', y='Plan_Length', label=key, logy=True)#, c=key, label=key)
plt.legend(bbox_to_anchor=(1,1))
```
